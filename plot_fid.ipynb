```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the csv file
file_path = 'train_dist_celeba.csv'
df = pd.read_csv(file_path)

# Define the plot size and style
plt.figure(figsize=(10, 6))
plt.style.use('seaborn-whitegrid')  # White background with gridlines for clarity

# Customize color and style of the plot lines
plt.plot(df['Step'], df['no_dist'], color='b', linewidth=2.5, label='No Dist')
plt.plot(df['Step'], df['dist'], color='r', linewidth=2.5, label='Dist')

# Customize titles and labels
plt.title("FID Scores over Steps", loc='center', fontsize=14, fontweight='bold')  # Centered title without coloring
plt.xlabel("Steps", fontsize=12, fontweight='bold')
plt.ylabel("FID Score", fontsize=12, fontweight='bold')

# Customize legend
plt.legend(loc='upper right', fontsize=10, frameon=True, shadow=True)

# Remove unnecessary spines for cleaner look
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the csv files
df1 = pd.read_csv('train_dist_celeba.csv')
df2 = pd.read_csv('train_dist_lsun_bed.csv')
df3 = pd.read_csv('train_dist_lsun_church.csv')

# Define the plot size and style
plt.figure(figsize=(30, 8))
plt.style.use('seaborn-whitegrid')

# Calculate global y-limits for consistency across subplots
y_min = min(df1['no_dist'].min(), df1['dist'].min(), df2['nodist'].min(), df2['dist'].min(), df3['nodist'].min(), df3['dist'].min())
y_max = max(df1['no_dist'].max(), df1['dist'].max(), df2['nodist'].max(), df2['dist'].max(), df3['nodist'].max(), df3['dist'].max())

# First plot (a)
plt.subplot(1, 3, 1)
plt.plot(df1['Step'], df1['no_dist'], color='b', linewidth=2.5, label='No Dist')
plt.plot(df1['Step'], df1['dist'], color='r', linewidth=2.5, label='Dist')
plt.title("a) CelebA Dataset", loc='center', fontsize=20, fontweight='bold')
plt.xlabel("Steps", fontsize=18, fontweight='bold')
plt.ylabel("FID Score", fontsize=18, fontweight='bold')  # Only this plot will have y-axis label
plt.ylim(y_min, y_max)  # Set y-limits
plt.legend(loc='upper right', fontsize=18, frameon=True, shadow=True)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().tick_params(axis='both', labelsize=16)

# Second plot (b)
plt.subplot(1, 3, 2)
plt.plot(df2['Step'], df2['nodist'], color='b', linewidth=2.5, label='No Dist')
plt.plot(df2['Step'], df2['dist'], color='r', linewidth=2.5, label='Dist')
plt.title("b) LSUN-Bedroom Dataset", loc='center', fontsize=20, fontweight='bold')
plt.xlabel("Steps", fontsize=18, fontweight='bold')
plt.ylim(y_min, y_max)  # Set y-limits
plt.legend(loc='upper right', fontsize=18, frameon=True, shadow=True)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().tick_params(axis='both', labelsize=16)

# Third plot (c)
plt.subplot(1, 3, 3)
plt.plot(df3['Step'], df3['nodist'], color='b', linewidth=2.5, label='No Dist')
plt.plot(df3['Step'], df3['dist'], color='r', linewidth=2.5, label='Dist')
plt.title("c) LSUN-Church Dataset", loc='center', fontsize=20, fontweight='bold')
plt.xlabel("Steps", fontsize=18, fontweight='bold')
plt.ylim(y_min, y_max)  # Set y-limits
plt.legend(loc='upper right', fontsize=18, frameon=True, shadow=True)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().tick_params(axis='both', labelsize=16)

# Adjust layout and show plots
plt.tight_layout()
plt.show()
```

```python
steps = 512
cumulative_updates = 0
updates_per_halving = 2000

while steps > 0:
    steps //= 2  # Halve the steps immediately
    updates_per_step = 20000 if steps in [1, 2, 4] else updates_per_halving
    updates = updates_per_step
    cumulative_updates += updates
    
    print(f"Steps: {steps}, Gradient Updates: {updates}, Cumulative Updates: {cumulative_updates}")
    
    if steps == 1:
        break  # Break the loop if steps reach 1 to avoid infinite loop
    
    #updates_per_halving += 2000  # Increment the updates per halving by 2000 at each step
```
